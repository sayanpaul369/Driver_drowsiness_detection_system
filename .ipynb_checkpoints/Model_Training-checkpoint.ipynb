```python
import tensorflow as tf
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dropout,Input,Flatten,Dense,MaxPooling2D
# from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
tf.test.is_gpu_available()
```

```python
train_datagen=ImageDataGenerator(rescale=1./255,
                                 rotation_range=0.2,
                                 shear_range=0.2,
                                 zoom_range=0.2,
                                 width_shift_range=0.2,
                                 height_shift_range=0.2,
                                 validation_split=0.2
                                )
```

```python
batchsize=8
```

```python
train_data = train_datagen.flow_from_directory(
    r'C:\Users\JOY KARMAKAR\Desktop\projects\driver_drowsiness_detection\Prepared_Data\Train',
    target_size=(80, 80),
    batch_size=batchsize,
    class_mode='categorical',
    subset='training'
)

validation_data= train_datagen.flow_from_directory(
    r'C:\Users\JOY KARMAKAR\Desktop\projects\driver_drowsiness_detection\Prepared_Data\Train',
    target_size=(80, 80),
    batch_size=batchsize,
    class_mode='categorical',
    subset='validation'
    
)
```

```python
test_datagen=ImageDataGenerator(rescale=1./255,)
test_data=test_datagen.flow_from_directory(
    r'C:\Users\JOY KARMAKAR\Desktop\projects\driver_drowsiness_detection\Prepared_Data\Test',
    target_size=(80, 80),
    batch_size=batchsize,
    class_mode='categorical'
)
```

```python
base_model=InceptionV3(include_top=False,weights='imagenet',input_tensor=Input(shape=(80,80,3)))
head_model=base_model.output
head_model=Flatten()(head_model)
head_model=Dense(64,activation='relu')(head_model)
head_model=Dropout(0.5)(head_model)
head_model=Dense(2,activation='softmax')(head_model)
model=Model(inputs=base_model.input,outputs=head_model)
for layer in base_model.layers:
    layer.trainable=False
    
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint,EarlyStopping,ReduceLROnPlateau
```

```python
checkpoint=ModelCheckpoint(r'C:\Users\JOY KARMAKAR\Desktop\projects\driver_drowsiness_detection\models\model.h5',
                          monitor='val_loss',
                           save_best_only=True,
                           verbose=3
                          )
earlystop=EarlyStopping(monitor='val_loss',patience=7,verbose=3,restore_best_weights=True)
learning_rate= ReduceLROnPlateau(monitor='val_loss',patience=3,verbose=3)
callbacks=[checkpoint,earlystop,learning_rate]

#In the learning process of neural networks, their weight parameters are updatediteratively so that the loss decreases.
#However, in some settings the loss does not decrease simply, butits decreasing speed slows down signicantly partway through learning, and then it speeds up againafter a long period of time.
#This is called as plateau phenomenon.





#we need to change the patience value according to the model accuracy . 
#if the accuracy and validation loss does not meet our expectations we need to increase the patience of the early stopping thus improving the accuracy.
#by setting the verbose we are motnitering the progress of the model training 
#verbose 3 means three parameters will be shown on the screen---- epoch number, progress bar, time
#Adam optimizer performs the best so we have selected this 
#in the checkpoint we are we are saving the best model according to the validation loss as model.h5
#if the learning rate is high it will be decreased according to the validation loss and plateu
```

```python
model.compile(optimizer='Adam',loss='categorical_crossentropy',metrics=['accuracy'])

model.fit_generator(train_data,steps_per_epoch=train_data.samples//batchsize,
                   validation_data=validation_data,
                   validation_steps=validation_data.samples//batchsize,
                   callbacks=callbacks,
                   epochs=50)
```

# Model evaluation

```python
accuracy_training,loss_training=model.evaluate_generator(train_data)
print(accuracy_training)
print(loss_training)
```

```python
accuracy_validation,loss_validation=model.evaluate_generator(validation_data)
print(accuracy_validation)
print(loss_validation)
```

```python
accuracy_test,loss_test=model.evaluate_generator(test_data)
print(accuracy_test)
print(loss_test)
```
